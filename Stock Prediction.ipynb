```python
# Import Libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
# Load and store Data
from google.colab import files
files.upload()
df = pd.read_csv('Reliance.csv')
```

```python
df.head()
```

```python
#Clean and Manipulate Data
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
```

```python
df = df.drop(columns = ['Date'])
```

```python
df = df.drop(columns = ['Average Price', 'Total Traded Quantity',	'Turnover',	'No. of Trades',	'Deliverable Qty',	'% Dly Qt to Traded Qty'])
```

```python
df.head()
```

```python
#Create Traget Column
df['Price_up'] = np.where(df['Close Price'].shift(-1) > df['Close Price'], 1,0)
```

```python
df = df.drop(columns = ['Symbol','Series'])
```

```python
df
```

```python
#Split Data
X = df.iloc[:, 0:df.shape[1]-1].values
Y = df.iloc[:,df.shape[1]-1].values
```

```python
#Split Data for Train and Test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
#Create and train model(DecisonTreeClassifier)
tree = DecisionTreeClassifier().fit(X_train, Y_train)

#Test Result
print(tree.score(X_test, Y_test))
```

```python
#Show Predictions
tree_predictions = tree.predict(X_test)
print(tree_predictions)
```
